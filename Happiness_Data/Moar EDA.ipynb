```python
# Load the csv
data = pd.read_csv('data_2017.csv')

# Initialize plot figure size
f,ax = plt.subplots(figsize = (14,8))

# Set title
title = "Correlation Matrix for Data 2017"

# Mask, this lets us display only half the matrix, since it is symmetric
mask = np.zeros_like(data.corr())
mask[np.triu_indices_from(mask)] = True

# Color styling
cmap = sns.diverging_palette(145, 280, s=85, l=25, n=20)

# Heatmap
t = sns.heatmap(data.corr(), annot = True, ax=ax, mask = mask, linewidths = 2, xticklabels=True, yticklabels=True).set_title(title)

# Set easier display sizes.
plt.yticks(rotation=0) 
plt.xticks(fontsize=9)
plt.yticks(fontsize=14)

# Display
plt.show()

t.figure.savefig("test.png")
```
